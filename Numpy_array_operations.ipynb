### What is Broadcasting? (Simple explanation)

Imagine you have a list of numbers, and you want to add a single number to each item in that list. Instead of writing code to go through each item one by one, NumPy's **broadcasting** lets you do this automatically.

It's like NumPy is smart enough to see that you want to add the single number to *every* element of the list, even though they don't look exactly the same shape. It *stretches* the single number (conceptually) to match the size of the list, then performs the operation.

### `np.hsplit()` (Horizontal Split)

*   `np.hsplit()` divides an array into multiple sub-arrays **horizontally**, meaning it cuts the array along its **columns**.
*   Imagine cutting a cake *vertically* through its layers to get slices side-by-side. Each slice still has all the layers.

### `np.vsplit()` (Vertical Split)

*   `np.vsplit()` divides an array into multiple sub-arrays **vertically**, meaning it cuts the array along its **rows**.
*   Imagine cutting a cake *horizontally* across its layers to separate the top half from the bottom half. Each piece is a row or group of rows.

**Key Difference:**
*   `hsplit` changes the **number of columns** in the resulting sub-arrays.
*   `vsplit` changes the **number of rows** in the resulting sub-arrays.

```python
import numpy as np

# Let's re-create our example array 'a' for clarity
a = np.arange(16).reshape(4,4)
print("Original Array 'a':\n", a)
print("--------------------\n")

# Using np.hsplit(a, 2) - Splits 'a' into 2 sub-arrays horizontally (column-wise)
print("Result of np.hsplit(a, 2) - Each sub-array has fewer columns:\n")
h_splits = np.hsplit(a, 2)
for i, sub_array in enumerate(h_splits):
    print(f"Horizontal Split {i+1}:\n", sub_array)
print("--------------------\n")

# Using np.vsplit(a, 2) - Splits 'a' into 2 sub-arrays vertically (row-wise)
# (changed from 4 to 2 for a clearer example of separating rows into chunks)
print("Result of np.vsplit(a, 2) - Each sub-array has fewer rows:\n")
v_splits = np.vsplit(a, 2)
for i, sub_array in enumerate(v_splits):
    print(f"Vertical Split {i+1}:\n", sub_array)
print("--------------------\n")
```
