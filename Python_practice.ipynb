```python
# 5.Count frequency of each element in an array
def num_frequency(arr):
    result={}
    for item in arr:
        if item in result:
            result[item]=result[item]+1
        else:
            result[item]=1
    return result


print(num_frequency([12,32,34,45,12,34,45,54,54,56,67,67,78,65,65,33,34]))
```
