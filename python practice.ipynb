sentence = "the quick brown fox jumps over the lazy dog"
Ek dictionary banao jisme har word ki frequency ho — but sirf un words ki jo 3 se zyada characters ke hon.

Expected: {'quick': 1, 'brown': 1, 'jumps': 1, 'over': 1, 'lazy': 1}

````python
def character_count(sentence):
    words={}
    counting=sentence.split()
    for item in counting:
        if len(item)>3:
            words[item]=words.get(item,0)+1
        
    return words
print(character_count("the quick brown fox jumps over the lazy dog"))
````

Large lists pe is approach ki kya problem ho sakti hai?```

````python
def mastery(lst):
    seen=set()
    result=[]
    for item in lst:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return list(set(result))
print(mastery([3,1,2,1,3,4]))
````

Q1. Tum ek list of strings ke paas ho:
pythonwords = ["  hello  ", "WORLD", "  python  ", "DATA"]
Ek single line mein aise list banao jisme har word stripped + lowercase ho.

Expected: ['hello', 'world', 'python', 'data']

````python
def lowered(strings):
    lst=[string.strip().lower() for string in strings ]
    return lst
print(lowered(["  hello  ", "WORLD", "  python  ", "DATA"]))
````
