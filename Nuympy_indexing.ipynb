```python
import numpy as np
```

**Numpy indexing and slicing of vectors**

```python
arr=np.arange(11,21)
arr
```

```python
arr_slice=arr[3::2]
arr_slice
```

**Numpy indexingand slicing of matrix**

```python
arr=np.arange(1,37).reshape(6,6)
arr
```

```python
arr[4:6,4:6]
```

```python
arr[:,3]
```

**Boolean Indexing**

```python
arr=np.arange(11,21)
arr
```

```python
bool_index=arr%2==0
bool_index
```

```python
arr=arr[bool_index]
arr
```
