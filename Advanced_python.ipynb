```python
def decorate(func):
    def wrapper(*args):
        
        print("calculating....")
        func(*args)
        print("Thank you")
    return wrapper

@decorate
def addition(*args):
    sum=0
    for i in args:
        sum=sum+i
    print(sum)
addition(12,27,45,76)
```
