## 🟢 Q5 [Tuple + Set]

```python
data = [
    (1, 'Alice', 'Data Science'),
    (2, 'Bob', 'Web Dev'),
    (3, 'Alice', 'ML'),
    (4, 'Charlie', 'Data Science'),
    (5, 'Bob', 'Data Science')
]
```

### Task

1. Unique students ke naam nikalo (set use karke)
2. Unique courses nikalo
3. Kitne students Data Science mein hain?

````python
data = [
    (1, 'Alice', 'Data Science'),
    (2, 'Bob', 'Web Dev'),
    (3, 'Alice', 'ML'),
    (4, 'Charlie', 'Data Science'),
    (5, 'Bob', 'Data Science')
]
students=set()
course=set()
count=0
for item in data:
    students.add(item[1])
    course.add(item[2])
    if item[2]=="Data Science":
        count+=1
print("Unique Students : ",students)
print("Unique Courses : ",course)
print("Data Science Students : ",count)

````

````python
data = [
    (1, 'Alice', 'Data Science'),
    (2, 'Bob', 'Web Dev'),
    (3, 'Alice', 'ML'),
    (4, 'Charlie', 'Data Science'),
    (5, 'Bob', 'Data Science')
]
students=set(students.add(item[1]) for item in data)
course=set(course.add(item[2]=="Data Science") for item in data)
count=0
for item in data:
    if item[2]=="Data Science":
        count+=1
print("Unique Students : ",students)
print("Unique Courses : ",course)
print("Data Science Students : ",count)

````
