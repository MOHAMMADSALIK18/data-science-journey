````python
def remove_duplicates(data):
    seen = []
    result = []
    for item in data:
        if item not in seen:
            result.append(item)
            seen.append(item)
    return result

print(remove_duplicates([1,2,2,3,3,4]))
````

# Q7. Word Frequency Counter

## Problem

Ek function `word_frequency(sentence)` banao jo ek sentence le aur har word ki frequency count karke dictionary return kare.

### Example Input

```python
"apple banana apple cherry banana apple"

````python
def word_frequency(sentence):
    result={}
    counting=(sentence.split())
    for item in counting:
        if item in result:
            result[item]=result[item]+1
        else:
            result[item]=1
    return result
print(word_frequency("apple banana apple cherry banana apple"))
````
