```python
class Factory:
  def __init__(self,materials,zips,pockets):
    self.materials=materials
    self.zips=zips
    self.pockets=pockets
    print(self)

  def show(self):
    print(f"Your requirements are:- {self.materials}.{self.zips},{self.pockets}")

reebock=Factory("Leathers",3,2)

campus=Factory("Nylon",4,3)

print(reebock.materials)
print(campus.zips)

reebock.show()
campus.show()

@classmethod
def hekko(cls):
  print("How are you.")

@staticmethod
def static():
  print("How are you brother.")

obj=Factory("Linen",1,4)
obj.show()
```
