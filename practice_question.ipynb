Sirf 'a' se shuru hone wale words count karo aur unhe alphabetically sorted print karo.

````python
words = ["apple", "banana", "avocado", "cherry", "apricot", "blueberry"]

# 'a' se start hone wale words
a_words = [word for word in words if word.startswith('a')]

# Alphabetically sort karna
a_words.sort()

# Count aur sorted words print karna
print("Count:", len(a_words))
print("Sorted Words:", a_words)
````

````python
sentence = "Data Science is Amazing and Data is Powerful"

# Case-insensitive aur spaces ignore
sentence = sentence.lower().replace(" ", "")

freq = {}

for char in sentence:
    freq[char] = freq.get(char, 0) + 1

# Maximum frequency wala character
max_char = max(freq, key=freq.get)

print("Most Frequent Character:", max_char)
print("Frequency:", freq[max_char])
````
