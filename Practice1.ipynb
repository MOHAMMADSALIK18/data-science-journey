##### Concept: String methods, loops, conditionals
###### Ek data cleaning task — raw sensor data aaya hai:
`pythonreadings = ["  38.5C  ", "INVALID", "  39.1C  ", "ERROR", "  37.8C  ", "  40.2C  ", "NULL"]`
##### Task:

###### Valid readings identify karo (jo "C" se end hon)
###### Strip karo whitespace
###### Sirf numeric part extract karo (e.g. "38.5")
###### Final clean list print karo as floats

````python
value= ["  38.5C  ", "INVALID", "  39.1C  ", "ERROR", "  37.8C  ", "  40.2C  ", "NULL"]
````

````python
clean=[]
for i in value:
    stripping=i.strip()
    if stripping.endswith("C"):
        num=stripping[:-1]
        clean.append(float(num))
print(clean)
````

#### Task:

###### Ek dictionary banao jisme har fruit ka count ho

###### Koi bhi built-in Counter use mat karo — manually karo

````python
fruits = ["apple", "banana", "apple", "orange", "banana", "apple", "mango", "banana"]
fruit={}
for item in fruits:
    if item in fruit:
        fruit[item]=fruit[item] + 1
    else:
        fruit[item]=1
print("fruit=",fruit)
        
````

````python
max_fruit=""
max_count=0
for k,v in fruit.items():
    if v>max_count:
        max_count=v
        max_fruit=k
print(max_fruit,max_count)
````

##### Q3 [Dictionary]

##### Ek list of words di gayi hai:

``words = ["apple", "banana", "avocado", "blueberry", "cherry", "apricot"]``

##### Ek dictionary banao jisme key = first letter, value = us letter se shuru hone wale saare words ki list.

````python
words = ["apple", "banana", "avocado", "blueberry", "cherry", "apricot"]
fruits={}
for item in words:
    letter=item[0]
    if letter not in fruits:
        fruits[letter]=[]
    fruits[letter].append(item)
print(fruits)
    
````

**Q4** `[Loop + Logic]` 

Real-world: **Temperature Converter**

```python
temps_celsius = [0, 20, 37, 100, -40, 25]
```

> Fahrenheit mein convert karo using **list comprehension**.

Formula:

```python
F = (C × 9/5) + 32
```

> Also tag karo:

* `"Cold"` if F < 50
* `"Warm"` if F is between 50–90
* `"Hot"` if F > 90

**Pehle khud solve karo.**
Hint: Is question mein tumhe **List Comprehension + Loop Logic + If-Else** tino ka use karna padega. 🚀


````python
temps_celsius = [0, 20, 37, 100, -40, 25]
temp_fahrenheit=[(temp*9/5)+32 for temp in temps_celsius]
for F in temp_fahrenheit: 
    if F<50:
        print("COLD")
    elif F>50 and F<90:
        print("WARM")
    else:
        print("HOT")
print(temp_fahrenheit)
````
