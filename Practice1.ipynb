
Q1 — 🟢 Easy | Loops + List
Concept: for loop, list traversal, conditional logic
Tumhare paas ek dataset hai — students ke marks:
`pythonmarks = [45, 82, 33, 91, 67, 55, 78, 29, 88, 61]
Task:`

Count karo kitne students ne pass kiya (marks ≥ 50)
Unka average nikalo (sirf passing students ka)
Highest aur lowest mark print karo

Built-in max(), min() allowed hain, lekin average khud calculate karo.

```python
pythonmarks = [45, 82, 33, 91, 67, 55, 78, 29, 88, 61]
```

```python
for i in pythonmarks:
    if i>50:
        print(i)
```

```python
passing=[]
for i in pythonmarks:
    if i>=50:
        passing.append(i)
print(passing)
```

```python
avg=sum(passing)/7
print(avg)
```

```python

print(min(pythonmarks))
print(max(pythonmarks))
```

Q2 — 🟢 Easy | String Manipulation
##### Concept: String methods, loops, conditionals
###### Ek data cleaning task — raw sensor data aaya hai:
`pythonreadings = ["  38.5C  ", "INVALID", "  39.1C  ", "ERROR", "  37.8C  ", "  40.2C  ", "NULL"]`
##### Task:

###### Valid readings identify karo (jo "C" se end hon)
###### Strip karo whitespace
###### Sirf numeric part extract karo (e.g. "38.5")
###### Final clean list print karo as floats

```python
value= ["  38.5C  ", "INVALID", "  39.1C  ", "ERROR", "  37.8C  ", "  40.2C  ", "NULL"]
```

```python
clean=[]
for i in value:
    stripping=i.strip()
    if stripping.endswith("C"):
        num=stripping[:-1]
        clean.append(float(num))
print(clean)
```

Q3 — 🟡 Medium | Dictionary
#### Concept: Dict creation, iteration, value operations
##### Ek small inventory system banao:
`pythonproducts = ["apple", "banana", "apple", "orange", "banana", "apple", "mango", "banana"]`
#### Task:

###### Ek dictionary banao jisme har fruit ka count ho
###### Sabse zyada bikne wala fruit print karo
###### Koi bhi built-in Counter use mat karo — manually karo

```python
fruits = ["apple", "banana", "apple", "orange", "banana", "apple", "mango", "banana"]
fruit={}
for item in fruits:
    if item in fruit:
        fruit[item]=fruit[item] + 1
    else:
        fruit[item]=1
print("fruit=",fruit)
        
```

```python
max_fruit=""
max_count=0
for k,v in fruit.items():
    if v>max_count:
        max_count=v
        max_fruit=k
print(max_fruit,max_count)
```
